### 4. Missing Values
##### 4.1 Categorical Variables

```python
# first let's filter out categorical features only as a list
cat_vars = [var for var in df.select_dtypes("object").columns]
cat_vars
```

```python
# number of categorical variables
len(cat_vars)
```

```python
# now let's capture all categorical variables with null values
cat_vars_with_na = [var for var in cat_vars if df[var].isnull().sum() > 0 ]
cat_vars_with_na
```

```python
# number of categorical variables with null values
len(cat_vars_with_na)
```

```python
# now let's see how much percent of data is missing in those categorical columns
df[cat_vars_with_na].isnull().mean().sort_values(ascending = False)
```

from the above code result we can see that the data frame have 16 categorical features with missing values.

##### Relationship Between missing values and target variable

```python
# let's analys the relationship between median of sale price for  missing and non missing records of each feature.
# let's define a function
def analyse_cat_missing(data, var):
    data = data.copy()
    
    data["var_flag"] = np.where(df[var].isnull(),1,0)
    data.groupby("var_flag")["SalePrice"].median().plot.bar()
    plt.title(var + "_Sale Price Comparison")
    plt.show()
    
for var in cat_vars_with_na:
    analyse_cat_missing(df,var)
```
