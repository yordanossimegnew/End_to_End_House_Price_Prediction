### 4. Missing Values
##### 4.1 Categorical Variables

```python
# first let's filter out categorical features only as a list
cat_vars = [var for var in df.select_dtypes("object").columns]
cat_vars
```

```python
# number of categorical variables
len(cat_vars)
```

```python
# now let's capture all categorical variables with null values
cat_vars_with_na = [var for var in cat_vars if df[var].isnull().sum() > 0 ]
cat_vars_with_na
```

```python
# number of categorical variables with null values
len(cat_vars_with_na)
```

```python
# now let's see how much percent of data is missing in those categorical columns
df[cat_vars_with_na].isnull().mean().sort_values(ascending = False)
```

from the above code result we can see that the data frame have 16 categorical features with missing values.

##### Relationship Between missing values and target variable

```python
# let's analys the relationship between median of sale price for  missing and non missing records of each feature.
# let's define a function
def analyse_cat_missing(data, var):
    data = data.copy()
    
    data["var_flag"] = np.where(df[var].isnull(),1,0)
    data.groupby("var_flag")["SalePrice"].median().plot.bar()
    plt.title(var + "_Sale Price Comparison")
    plt.ylabel("Median Sale Price")
    plt.xlabel("Missing Information Flag")
    
    out_put_dir = "C:\\Users\\yozil\\Desktop\\My projects\\9.0 End_to_End_House_Price_Prediction\\reports\\figures\\Median Sale Price Missing vs Non-Missing"
    
    plt.savefig(os.path.join(out_put_dir,f"{var}_sale_price_comparison.jpg"))
    plt.show()
    
for var in cat_vars_with_na:
    analyse_cat_missing(df,var)
```

The Average Sale Price in houses where the information is missing, differs from the average Sale Price in houses where information exists.

##### 4.2.1. Discrete Variables

```python
# first let's filter out the discrete varaibles as a list
disc_vars = [var for var in num_vars if df[var].nunique()<20 ]
disc_vars
```

```python
# Number of discrete variables
len(disc_vars)
```

```python
# now let's explore the existance of missing values in those discrete variable columns
df[disc_vars].isnull().mean()
```

As we can see from the above result we don't have any missing values in the discrete variables.

```python
# now let's see the median sale price for the values in each discrete varaibles
# let's start by defining a function
def analyse_discrete(data, var):
    data = data.copy()
    
    data.groupby(var)["SalePrice"].median().plot.bar()
    plt.title(var + "_Sale-Price_Analysis")
    plt.xlabel(var)
    plt.ylabel("Sale Price")
    outupt_path  = "C:\\Users\\yozil\\Desktop\\My projects\\9.0 End_to_End_House_Price_Prediction\\reports\\figures\\sale price for discrete variables"
    plt.savefig(os.path.join(outupt_path,f"{var}_sale price analysis.jpg"))
    plt.show()
    
for var in disc_vars:
    analyse_discrete(df,  var)
```

There tend to be a relationship between the variables values and the SalePrice, but this relationship is not always monotonic. 

For example, for OverallQual, there is a monotonic relationship: the higher the quality, the higher the SalePrice.  

However, for OverallCond, the relationship is not monotonic. Clearly, some Condition grades, like 5, correlate with higher sale prices, but higher values do not necessarily do so. We need to be careful on how we engineer these variables to extract maximum value for a linear model.

##### Distribution of Continous Variables

```python
# Now let's visualize the distribution of continous variables.
# let's define a function
def analyse_continous(data, var):
    data = data.copy()
    
    data[var].plot.hist()
    plt.title(var+ "_Distribution")
    plt.xlabel(var)
    plt.ylabel("Frequency_of_" + var)
    
    output_path = "C:\\Users\\yozil\\Desktop\\My projects\\9.0 End_to_End_House_Price_Prediction\\reports\\figures\\continous_variables_distribution"
    
    plt.savefig(os.path.join(output_path, f"_{var}_distribution.jpg"))
    plt.show()
    
for var in cont_vars:
    analyse_continous(df,var)
```
