##### Relationship Between missing values and target variable

```python
# let's analys the relationship between median of sale price for  missing and non missing records of each feature.
# let's define a function
def analyse_cat_missing(data, var):
    data = data.copy()
    
    data["var_flag"] = np.where(df[var].isnull(),1,0)
    data.groupby("var_flag")["SalePrice"].median().plot.bar()
    plt.title(var + "_Sale Price Comparison")
    plt.ylabel("Median Sale Price")
    plt.xlabel("Missing Information Flag")
    
    out_put_dir = "C:\\Users\\yozil\\Desktop\\My projects\\9.0 End_to_End_House_Price_Prediction\\reports\\figures\\Median Sale Price Missing vs Non-Missing"
    
    plt.savefig(os.path.join(out_put_dir,f"{var}_sale_price_comparison.jpg"))
    plt.show()
    
for var in cat_vars_with_na:
    analyse_cat_missing(df,var)
```

The Average Sale Price in houses where the information is missing, differs from the average Sale Price in houses where information exists.

##### Rare Labels

```python
# Let's go ahead and investigate now if there are labels that are present only in a small number of houses:
# Let's start by defining a function
def analyse_rare(data, var, rare_percent):
    data = data.copy()
    
    temp = data.groupby(var)[var].count()/len(df)
    
    return temp[temp < rare_percent]

for var in cat_vars:
    print(analyse_rare(df, var, rare_percent= 0.01))
    print()
```

Some of the categorical variables show multiple labels that are present in less than 1% of the houses.

```python
# relationship between categorical variables and the target_column "Sale Price"

def analyse_cat_vs_SalePrice(data, var):
    data = data.copy()
    
    data.groupby(var)["SalePrice"].median().plot.bar()
    plt.xlabel(var)
    plt.ylabel("Sale Price")
    plt.title(f"{var}_vs Sale_Price")
    
    output_path ="C:\\Users\\yozil\\Desktop\\My projects\\9.0 End_to_End_House_Price_Prediction\\reports\\figures\\cat_vars_vs_sale_price"
    plt.savefig(os.path.join(output_path,f"{var}_vs_saleprice.jpg"))
    plt.show()
    
for var in cat_vars:
    analyse_cat_vs_SalePrice(df, var)
```

Clearly, the categories give information on the SalePrice, as different categories show different median sale prices.

##### Distribution of Continous Variables

```python
# Now let's visualize the distribution of continous variables.
# let's define a function
def analyse_continous(data, var):
    data = data.copy()
    
    data[var].plot.hist()
    plt.title(var+ "_Distribution")
    plt.xlabel(var)
    plt.ylabel("Frequency_of_" + var)
    
    output_path = "C:\\Users\\yozil\\Desktop\\My projects\\9.0 End_to_End_House_Price_Prediction\\reports\\figures\\continous_variables_distribution"
    
    plt.savefig(os.path.join(output_path, f"_{var}_distribution.jpg"))
    plt.show()
    
for var in cont_vars:
    analyse_continous(df,var)
```

The variables are not normally distributed, including the target variable 'SalePrice'. 

To maximise performance of linear models, we need to account for non-Gaussian distributions.

Let's evaluate if a logarithmic transformation of the variables returns values that follow a normal distribution:

##### Gaussian Transformation

we apply gaussian transformation for all variables with values > 0

```python
# first let's mask those columns with only positive values as a list
trans_vars = [var for var in cont_vars if all(df[var]) > 0]
trans_vars
```

As shown above we have 5 continous features that have a value all greater than 0, so let's apply the gaussian transformation on this features

```python
def analyse_transform_dist(data, var):
    data = data.copy()
    
    np.log(data[var]).plot.hist()
    plt.title(var + "_transform_distribution")
    plt.xlabel(var)
    plt.ylabel("Frequency")
    
    output_path = "C:\\Users\\yozil\\Desktop\\My projects\\9.0 End_to_End_House_Price_Prediction\\reports\\figures\\transformed_continous_variables_distribution"
    plt.savefig(os.path.join(output_path,f"{var}_transformed_distribution.jpg"))
    
    plt.show()
    
for var in trans_vars:
    analyse_transform_dist(df,var)
```

We get a better spread of the values for most variables when we use the logarithmic transformation. This engineering step will most likely add performance value to our final model.

```python
# Now Let's Explore the relationship between house sale price and the transformed varaibales
def analyse_tranformed_vs_saleprice(data, var):
    data = data.copy()
    
    sns.scatterplot(data = data, x= var, y= "SalePrice")
    plt.xlabel("SalePrice")
    plt.ylabel(var)
    plt.title(f"Sale Price vs {var} ")
    
    output_path = "C:\\Users\\yozil\\Desktop\\My projects\\9.0 End_to_End_House_Price_Prediction\\reports\\figures\\sale_pirce_vs_transformed_variables"
    plt.savefig(os.path.join(output_path,f"{var} vs Sale Price.jpg"))
    
    plt.show()
    
    
for var in trans_vars:
    if var != "SalePrice":
        analyse_tranformed_vs_saleprice(df, var)
```

From the previous plots, we observe some monotonic associations between SalePrice and the variables to which we applied the log transformation, for example 'GrLivArea'.

##### Outliers

Extreme values may affect the performance of a linear model. Let's find out if we have any in our variables.

```python
# let's plot the distribution of the transformed variables in box plot to see outliers
def analyse_outliers(data, var):
    data = data.copy()
    
    data[var].plot.box()
    plt.title(f"distribution of {var}")
    
    output_path = "C:\\Users\\yozil\\Desktop\\My projects\\9.0 End_to_End_House_Price_Prediction\\reports\\figures\\transformed_vars_ouliers"
    plt.savefig(os.path.join(output_path,f"{var}_distribution.jpg"))
    plt.show()
    
for var in trans_vars:
    analyse_outliers(df, var)
```

The majority of the continuous variables seem to contain outliers. Outliers tend to affect the performance of linear model. So it is worth spending some time understanding if removing outliers will add performance value to our  final machine learning model.

#### Temporal variables

We have 4 year variables in the dataset:

- YearBuilt: year in which the house was built
- YearRemodAdd: year in which the house was remodeled
- GarageYrBlt: year in which a garage was built
- YrSold: year in which the house was sold

We generally don't use date variables in their raw format. Instead, we extract information from them. For example, we can capture the difference in years between the year the house was built and the year the house was sold.

```python
# the temporal variables are
year_vars
```

```python
# Now let's see unique values in the temporal varaibles
for var in year_vars:
    print(df[var].unique())
    print()
```

```python
# first let's see how the salling price changes as the year that the house sold changes
df.groupby("YrSold")["SalePrice"].median().plot()
plt.xlabel("Year Sold")
plt.ylabel("Median Price")
plt.title("Selling Year vs Median Price")
plt.xticks(df["YrSold"].unique())
plt.savefig("C:\\Users\\yozil\\Desktop\\My projects\\9.0 End_to_End_House_Price_Prediction\\reports\\figures\\temporal_vars analysis\\Selling Year vs Median Price.jpg")
plt.show()
```

There has been a drop in the value of the houses. That is unusual, in real life, house prices typically go up as years go by.


Let's go ahead and explore whether there is a relationship between the year variables and SalePrice. For this, we will capture the elapsed years between the Year variables and the year in which the house was sold:

```python
# let's capture the year difference
def analyse_temporal(data, var):
    data = data.copy()
    data["Diff"] = data["YrSold"] - data[var]
    plt.scatter(data["Diff"], data["SalePrice"])
    plt.xlabel("Years")
    plt.ylabel("Sale Price")
    plt.title(f"{var}_vs_Sale_Price")
    output_path = "C:\\Users\\yozil\\Desktop\\My projects\\9.0 End_to_End_House_Price_Prediction\\reports\\figures\\temporal_vars analysis"
    plt.savefig(os.path.join(output_path,f"{var}_vs_sale_price.jpg"))
    plt.show()

for var in year_vars:
    if var != "YrSold":
        analyse_temporal(df, var)
    
    
```
