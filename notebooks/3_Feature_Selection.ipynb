## 2. Feature Selection

Let's go ahead and select a subset of the most predictive features. There is an element of randomness in the Lasso regression.

```python
# initialize the select from model class
sel = SelectFromModel(estimator = Lasso(alpha=0.001, random_state=0) )
# fit the select from model class to the training set.
sel.fit(x_train, y_train)
```

```python
# let's see which are the selected features
features = sel.get_support()
features
```

```python
# let's display the name of the columns
selected_features = x_train.columns[(features)]
selected_features
```

```python
# now let's see the number of selected features
len(selected_features)
```

Out of 81 features I have selected 36 features for building the model.
